```python
# train whole dataset
model_path = '/mnt/qb/work/mlcolab/hzhou52/kt/logs/TestHSSM/junyi15/multi_skill/2023-04-12T20:35:53.846125__overfit_0/Model/Model_20.pt'
state_dict = torch.load(model_path)

node_dict = {}
for key, value in state_dict.items():
    if 'node_' in key:
        node_dict[key] = value
node_u = node_dict['node_dist.u']
query_w = node_dict['node_dist.query_layer.weight'] 
query_b = node_dict['node_dist.query_layer.bias']
key_w = node_dict['node_dist.key_layer.weight']
key_b = node_dict['node_dist.key_layer.bias']

query = node_u @ query_w.transpose(0,1) + query_b.reshape(1,-1)
key = node_u @ key_w.transpose(0,1) + key_b.reshape(1,-1)

node_u = node_u.detach().cpu().numpy()
query = query.detach().cpu().numpy()
key = key.detach().cpu().numpy()
```

```python
# exercise-id re-match
base_pth = '/mnt/qb/work/mlcolab/hzhou52/kt/junyi15/'

exercise = pd.read_csv(base_pth+'junyi_Exercise_table.csv', encoding = "utf-8",low_memory=False) 

num_ex = len(exercise)

ex_new_id = pd.DataFrame(columns=['exercise_name', 'exercise_id', 'prerequisite_name', 'prerequisite_id'])
ex_new_id['exercise_name'] = exercise.name
ex_new_id['exercise_id'] = np.arange(len(exercise))
ex_new_id['prerequisite_name'] = exercise.prerequisites
ex_new_id['topic'] = exercise.topic
ex_new_id['area'] = exercise.area

for i in range(len(ex_new_id.prerequisite_name)):
    pre = ex_new_id['prerequisite_name'][i]
    ind = exercise.loc[exercise.name == pre].index
    if len(ind) == 0:
        ex_new_id.prerequisite_id[i] = np.nan
    else:
        ex_new_id.prerequisite_id[i] = ind[0]
```

```python
import networkx as nx
G = nx.DiGraph()

# create nodes for the graph, with label as the nodes' name
node_label = [(ex_new_id.exercise_id[i], dict(name=ex_new_id.exercise_name[i])) for i in range(len(ex_new_id))]
G.add_nodes_from(node_label)

# the edge of the graph; a->b means a is a prerequiste of b
edge = []
for i in range(len(ex_new_id)):
    if not pd.isna(ex_new_id.prerequisite_id[i]):
        edge.append((ex_new_id.prerequisite_id[i], ex_new_id.exercise_id[i]))
G.add_edges_from(edge)

# add position of each node on the graph
pos = dict()
for i in range(num_ex):
    pos.update({i: (exercise.h_position[i], exercise.v_position[i])})

ex_new_id.loc[ex_new_id.area.isna(), 'area'] = 'Dont_Know'
ex_new_id['area_id'] = pd.Categorical(ex_new_id['area'], categories=ex_new_id['area'].unique()).codes
```
