```python
import numpy as np
import math

def sigmoid(x, alpha=1):
  return 1 / (1 + alpha*np.exp(-x))
```

```python
u = np.random.normal(size=(10,4))
v = np.random.normal(size=(10,4))
```

```python
u
```

```python
adj = sigmoid(np.dot(u, v.transpose(-1,-2)), alpha=5)
adj = adj * (1-np.eye(10))
(adj>=0.5).sum()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

edges = []
for i in range(10):
    for j in range(10):
        if adj[i, j] >= 0.5:
            edges.append([i,j])
G = nx.DiGraph()
G.add_edges_from(edges)
```

```python
# explicitly set positions
pos = {0: (0, 1), 1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (1, 0.55), 5: (5, 0.8), 6:(6,0.1), 7:(7,0.3), 8:(8, 0.06),
          9:(9, 0.7)}
options = {
    "font_size": 15,
    "node_size": 1000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 2,
    "width": 2,
}
nx.draw_networkx(G, pos, **options)
```
